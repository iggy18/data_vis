```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
gold = pd.read_csv("StudentsPerformance.csv")
gold.head()
```

```python
print(gold.columns)
```

```python
sns.set_theme()
```

```python
sns.lmplot(x='math score',y='writing score',hue='gender',data=gold)
plt.xlabel('Math Score')
plt.ylabel('Writing Score')
plt.title('Math Score vs Writing Score')
plt.show()
```

## women get better test scores in math and writing on average

```python
sns.countplot(y=gold['parental level of education'],palette="Set3",hue=gold['gender'])
plt.legend(loc=4)
plt.show()
```

## if they graduate highschool women obtain more higher education degrees than men
